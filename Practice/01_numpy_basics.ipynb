```python
import numpy as np

#create 1D array
a=np.array([1,2,3,4,5])

print(a)
print()

#create 2D array
b=np.array([
    [1,2,3],
    [4,5,6],
    [6,7,8]
])

print(b)
```

```python
#matrix properties
import numpy as np

a=np.array([
    [2,4,6],
    [8,10,12],
    [67,89,29]
])
b=np.array(["shashwat","ajay","atul"])

print(a.ndim)
print(a.dtype)
print(b.dtype)
```

```python
#special matrix types
import numpy as np

a=np.zeros([3,3]) #2D ARRAY
print(a)
print()
b=np.eye(3) #identity matrix
print(b)
```

```python
#basic operations & statistics
import numpy as np

a=np.array([
    [1,2,3],
    [2,3,4],
    [3,4,5]
])

b=np.array([
    [4,5,6],
    [6,7,8],
    [7,8,9]
])

print(a+b)
print()
print(a*b) #element wise multiplication
print()
print(np.dot(a,b)) #matrix multiplication
print()
#basic stats operations
print(np.mean(a))
print(np.median(a))
print(np.var(a))
print(np.std(a))
```

CALCULATE AVG MAX MEAN MEDIAN OF THE MARKS

```python

import numpy as np

marks=np.array([
[85,90,88],
[70,75,80],
[95,92,96]
])

print(marks)

print(np.mean(marks))
print()
print(np.mean(marks,axis=1))#for each student(row-wise)
print()
print(np.mean(marks,axis=0))#for each subject(column-wise)
print()
print(np.max(marks))
print()
print(np.mean(marks,axis=1))

print()
#max avg of student
print(np.max(np.mean(marks,axis=1)))
```
